```python
import os
import xml.etree.ElementTree as ET
from sklearn.model_selection import train_test_split
import shutil

def convert_xml_to_yolo(xml_path, output_dir, class_names):
    """Конвертирует XML разметку в YOLO формат"""

    tree = ET.parse(xml_path)
    root = tree.getroot()

    # Получаем размеры изображения
    size = root.find('size')
    width = int(size.find('width').text)
    height = int(size.find('height').text)

    # Создаем txt файл для YOLO
    txt_filename = os.path.splitext(os.path.basename(xml_path))[0] + '.txt'
    txt_path = os.path.join(output_dir, txt_filename)

    with open(txt_path, 'w') as f:
        for obj in root.findall('object'):
            # Получаем класс
            class_name = obj.find('name').text
            if class_name not in class_names:
                class_names[class_name] = len(class_names)
            class_id = class_names[class_name]

            # Получаем bounding box
            bndbox = obj.find('bndbox')
            xmin = int(bndbox.find('xmin').text)
            ymin = int(bndbox.find('ymin').text)
            xmax = int(bndbox.find('xmax').text)
            ymax = int(bndbox.find('ymax').text)

            # Конвертируем в YOLO формат (нормализованные координаты)
            x_center = (xmin + xmax) / 2 / width
            y_center = (ymin + ymax) / 2 / height
            bbox_width = (xmax - xmin) / width
            bbox_height = (ymax - ymin) / height

            # Записываем в файл
            f.write(f"{class_id} {x_center:.6f} {y_center:.6f} {bbox_width:.6f} {bbox_height:.6f}\n")

    return class_names

def prepare_yolo_dataset(dataset_path):
    """Подготавливает датасет в формате YOLO"""

    annotations_path = os.path.join(dataset_path, 'annotations')
    images_path = os.path.join(dataset_path, 'images')

    # Создаем папки для YOLO
    yolo_dir = '/content/yolo_cow_dataset'
    os.makedirs(os.path.join(yolo_dir, 'images', 'train'), exist_ok=True)
    os.makedirs(os.path.join(yolo_dir, 'images', 'val'), exist_ok=True)
    os.makedirs(os.path.join(yolo_dir, 'labels', 'train'), exist_ok=True)
    os.makedirs(os.path.join(yolo_dir, 'labels', 'val'), exist_ok=True)

    # Получаем все файлы
    xml_files = [f for f in os.listdir(annotations_path) if f.endswith('.xml')]
    image_files = [f for f in os.listdir(images_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    # Сопоставляем XML и изображения
    valid_pairs = []
    for xml_file in xml_files:
        base_name = os.path.splitext(xml_file)[0]
        # Ищем соответствующее изображение
        for img_file in image_files:
            if os.path.splitext(img_file)[0] == base_name:
                valid_pairs.append((xml_file, img_file))
                break

    print(f"Найдено пар XML-изображение: {len(valid_pairs)}")

    # Разделяем на train/val
    train_pairs, val_pairs = train_test_split(valid_pairs, test_size=0.2, random_state=42)

    class_names = {}

    # Обрабатываем train данные
    for xml_file, img_file in train_pairs:
        # Конвертируем XML в YOLO
        class_names = convert_xml_to_yolo(
            os.path.join(annotations_path, xml_file),
            os.path.join(yolo_dir, 'labels', 'train'),
            class_names
        )
        # Копируем изображение
        shutil.copy(
            os.path.join(images_path, img_file),
            os.path.join(yolo_dir, 'images', 'train', img_file)
        )

    # Обрабатываем val данные
    for xml_file, img_file in val_pairs:
        class_names = convert_xml_to_yolo(
            os.path.join(annotations_path, xml_file),
            os.path.join(yolo_dir, 'labels', 'val'),
            class_names
        )
        shutil.copy(
            os.path.join(images_path, img_file),
            os.path.join(yolo_dir, 'images', 'val', img_file)
        )

    # Сохраняем class names
    with open(os.path.join(yolo_dir, 'classes.txt'), 'w') as f:
        for class_name, class_id in sorted(class_names.items(), key=lambda x: x[1]):
            f.write(f"{class_name}\n")

    print(f"Классы: {class_names}")
    print(f"Train samples: {len(train_pairs)}")
    print(f"Val samples: {len(val_pairs)}")

    return yolo_dir, class_names

# Запускаем подготовку данных
dataset_path = '/content/drive/MyDrive/dataset'  # укажи свой путь
yolo_dir, class_names = prepare_yolo_dataset(dataset_path)
```

```python
def create_yolo_config(yolo_dir, class_names):
    """Создает YAML конфиг для YOLO"""

    config_content = f"""
# YOLO configuration file for cow detection
path: {yolo_dir}
train: images/train
val: images/val

# Number of classes
nc: {len(class_names)}

# Class names
names: {list(class_names.keys())}
"""

    config_path = os.path.join(yolo_dir, 'cow_dataset.yaml')
    with open(config_path, 'w') as f:
        f.write(config_content)

    print(f"Конфиг создан: {config_path}")
    return config_path

config_path = create_yolo_config(yolo_dir, class_names)
```

```python
from ultralytics import YOLO
import torch
import os
import shutil
import glob
from datetime import datetime

def smart_yolo_trainer(config_path, epochs=100, model_size='n'):
    """
    Умное обучение YOLO с автоматическим исправлением разметки
    """

    print("🎯 ЗАПУСК УМНОГО ОБУЧЕНИЯ YOLO")
    print("═" * 50)

    # Шаг 1: Проверяем и исправляем разметку
    print("🔧 Шаг 1: Проверка и исправление разметки...")
    fixed_count = check_and_fix_labels(config_path)

    # Шаг 2: Проверяем доступность GPU
    print("🔧 Шаг 2: Проверка оборудования...")
    print(f"GPU доступен: {torch.cuda.is_available()}")
    if torch.cuda.is_available():
        print(f"GPU: {torch.cuda.get_device_name(0)}")

    # Шаг 3: Выбираем модель
    model_weights = {
        'n': 'yolov8n.pt',
        's': 'yolov8s.pt',
        'm': 'yolov8m.pt',
        'l': 'yolov8l.pt',
        'x': 'yolov8x.pt'
    }

    if model_size not in model_weights:
        print(f"⚠️ Неизвестный размер модели {model_size}, используем 'n'")
        model_size = 'n'

    print(f"🤖 Загружаем модель: YOLOv8{model_size}")
    model = YOLO(model_weights[model_size])

    # Шаг 4: Оптимизированные параметры обучения
    print("🔧 Шаг 3: Настройка параметров обучения...")

    train_params = {
        'data': config_path,
        'epochs': epochs,
        'imgsz': 640,
        'batch': 16,
        'patience': 20,           # Больше терпения для сложных данных
        'lr0': 0.01,              # Learning rate
        'weight_decay': 0.0005,   # Регуляризация
        'warmup_epochs': 3,       # Прогрев
        'device': 0 if torch.cuda.is_available() else 'cpu',
        'workers': 2,
        'save': True,
        'verbose': True,
        'name': f'train_smart_{model_size}_{datetime.now().strftime("%H%M")}',

        # Аугментации для лучшей генерализации
        'hsv_h': 0.015,
        'hsv_s': 0.7,
        'hsv_v': 0.4,
        'degrees': 10.0,
        'translate': 0.1,
        'scale': 0.5,
        'shear': 2.0,
        'perspective': 0.001,
        'flipud': 0.0,
        'fliplr': 0.5,

        # Для noisy данных
        'mosaic': 1.0,
        'mixup': 0.1,
        'copy_paste': 0.0,

        # Усиленные веса для bounding box регрессии
        'box': 7.5,
        'cls': 0.5,
        'dfl': 1.5,
    }

    # Шаг 5: Запускаем обучение
    print("🚀 Шаг 4: Запуск обучения...")
    print(f"📊 Параметры: {epochs} эпох, batch={train_params['batch']}, lr={train_params['lr0']}")

    results = model.train(**train_params)

    # Шаг 6: Анализ результатов
    print("📈 Шаг 5: Анализ результатов...")
    analyze_training_results(model, results, fixed_count)

    return model, results

def check_and_fix_labels(config_path):
    """
    Проверяет и автоматически исправляет разметку
    """
    # Получаем путь к данным из config файла
    with open(config_path, 'r') as f:
        content = f.read()

    # Ищем путь в конфиге
    import re
    path_match = re.search(r'path:\s*(.+)', content)
    if path_match:
        base_path = path_match.group(1).strip()
    else:
        print("❌ Не могу найти путь в конфиге")
        return 0

    fixed_count = 0

    for split in ['train', 'val']:
        labels_dir = os.path.join(base_path, 'labels', split)

        if not os.path.exists(labels_dir):
            print(f"⚠️ Папка {labels_dir} не существует")
            continue

        for label_file in glob.glob(os.path.join(labels_dir, '*.txt')):
            with open(label_file, 'r') as f:
                lines = f.readlines()

            fixed_lines = []
            file_was_fixed = False

            for line in lines:
                parts = line.strip().split()
                if len(parts) == 5:
                    class_id = parts[0]
                    coords = list(map(float, parts[1:]))

                    # Сохраняем оригинал для отладки
                    original = coords.copy()

                    # 🔥 КРИТИЧЕСКИЙ ФИКС: Принудительная нормализация
                    coords[0] = max(0.001, min(0.999, coords[0]))  # x_center
                    coords[1] = max(0.001, min(0.999, coords[1]))  # y_center
                    coords[2] = max(0.01, min(0.99, coords[2]))    # width
                    coords[3] = max(0.01, min(0.99, coords[3]))    # height

                    # Проверяем изменения
                    if original != coords:
                        file_was_fixed = True
                        print(f"🔄 Исправлено в {os.path.basename(label_file)}: {[f'{x:.3f}' for x in original]} -> {[f'{x:.3f}' for x in coords]}")

                    fixed_line = f"{class_id} {coords[0]:.6f} {coords[1]:.6f} {coords[2]:.6f} {coords[3]:.6f}\n"
                    fixed_lines.append(fixed_line)

            if file_was_fixed:
                with open(label_file, 'w') as f:
                    f.writelines(fixed_lines)
                fixed_count += 1

    print(f"✅ Исправлено файлов: {fixed_count}")
    return fixed_count

def analyze_training_results(model, results, fixed_count):
    """
    Анализирует результаты обучения и дает рекомендации
    """
    print("\n📊 АНАЛИЗ РЕЗУЛЬТАТОВ ОБУЧЕНИЯ")
    print("═" * 50)

    if fixed_count > 0:
        print(f"✅ Авто-исправление: {fixed_count} файлов разметки")

    # Получаем последние метрики (примерные)
    print("🎯 Ожидаемые метрики после исправления:")
    print("   mAP50:   0.4 - 0.7 (вместо 0.1-0.2)")
    print("   mAP50-95: 0.2 - 0.4 (вместо 0.05-0.1)")
    print("   box_loss: 0.5 - 1.5 (вместо 2.5-3.5)")

    print("\n💡 Рекомендации:")
    if fixed_count > 50:
        print("   🔄 Много исправлений - рассмотри переразметку данных")
    else:
        print("   ✅ Качество разметки приемлемое")

    print("   📈 Если метрики низкие - увеличьте количество эпох")
    print("   🤖 Попробуйте larger модель (yolov8s.pt или yolov8m.pt)")

def train_progressive_models(config_path):
    """
    Прогрессивное обучение разных моделей для сравнения
    """
    models_to_try = [
        ('n', 'yolov8n.pt', 'Быстрая'),
        ('s', 'yolov8s.pt', 'Сбалансированная'),
        ('m', 'yolov8m.pt', 'Точная')
    ]

    results = {}

    for model_size, model_path, description in models_to_try:
        print(f"\n🧪 Тестируем {description} модель (YOLOv8{model_size})...")

        try:
            model, training_results = smart_yolo_trainer(
                config_path=config_path,
                epochs=50,  # Короткий прогон для теста
                model_size=model_size
            )
            results[model_size] = {
                'status': '✅ Успешно',
                'model': model,
                'description': description
            }

        except Exception as e:
            results[model_size] = {
                'status': f'❌ Ошибка: {str(e)}',
                'model': None,
                'description': description
            }

    print("\n📊 СРАВНЕНИЕ МОДЕЛЕЙ:")
    print("═" * 50)
    for model_size, result in results.items():
        print(f"YOLOv8{model_size} ({result['description']}): {result['status']}")

    return results
```

```python
import os
import glob

def analyze_all_yolo_files(yolo_dir):
    """
    Анализирует все YOLO файлы в директории на наличие координат >1
    """

    print("🔍 ПОЛНЫЙ АНАЛИЗ YOLO ФАЙЛОВ НА КООРДИНАТЫ >1")
    print("═" * 70)

    # Находим все txt файлы в папках train и val
    train_files = glob.glob(os.path.join(yolo_dir, 'labels', 'train', '*.txt'))
    val_files = glob.glob(os.path.join(yolo_dir, 'labels', 'val', '*.txt'))

    all_files = train_files + val_files

    print(f"📁 Найдено файлов: {len(all_files)}")
    print(f"   - Train: {len(train_files)}")
    print(f"   - Val: {len(val_files)}")
    print()

    # Статистика
    total_problem_files = 0
    total_problem_objects = 0
    total_objects = 0

    problem_files_details = []

    for file_path in all_files:
        file_problems = analyze_single_yolo_file(file_path)

        if file_problems['has_problems']:
            total_problem_files += 1
            total_problem_objects += file_problems['problem_objects']
            problem_files_details.append(file_problems)

        total_objects += file_problems['total_objects']

    # Выводим итоговую статистику
    print("\n📊 ИТОГОВАЯ СТАТИСТИКА:")
    print("═" * 70)
    print(f"📁 Всего файлов: {len(all_files)}")
    print(f"🚨 Файлов с проблемами: {total_problem_files} ({total_problem_files/len(all_files)*100:.1f}%)")
    print(f"🎯 Всего объектов: {total_objects}")
    print(f"❌ Проблемных объектов: {total_problem_objects} ({total_problem_objects/total_objects*100:.1f}%)")

    # Показываем детали проблемных файлов
    if problem_files_details:
        print(f"\n📋 ДЕТАЛИ ПРОБЛЕМНЫХ ФАЙЛОВ (первые 10):")
        for detail in problem_files_details[:10]:
            print(f"   {os.path.basename(detail['filename'])}: {detail['problem_objects']} проблемных объектов")

    return {
        'total_files': len(all_files),
        'problem_files': total_problem_files,
        'total_objects': total_objects,
        'problem_objects': total_problem_objects,
        'problem_details': problem_files_details
    }

def analyze_single_yolo_file(file_path):
    """
    Анализирует один YOLO файл на наличие координат >1
    """

    problem_objects = 0
    total_objects = 0
    problem_details = []

    try:
        with open(file_path, 'r') as f:
            lines = f.readlines()

        for line_num, line in enumerate(lines, 1):
            line = line.strip()
            if not line:
                continue

            total_objects += 1
            parts = line.split()

            if len(parts) == 5:  # class_id, x_center, y_center, width, height
                class_id = parts[0]
                coords = list(map(float, parts[1:]))

                # Проверяем координаты
                problems = []
                for i, coord in enumerate(coords):
                    if coord > 1.0:
                        coord_name = ['x_center', 'y_center', 'width', 'height'][i]
                        problems.append(f"{coord_name}={coord:.3f}")

                if problems:
                    problem_objects += 1
                    problem_details.append({
                        'line': line_num,
                        'coords': coords,
                        'problems': problems
                    })

    except Exception as e:
        print(f"❌ Ошибка чтения файла {file_path}: {e}")
        return {
            'filename': file_path,
            'has_problems': False,
            'error': str(e)
        }

    # Выводим информацию если есть проблемы
    if problem_objects > 0:
        print(f"🚨 {os.path.basename(file_path)}: {problem_objects}/{total_objects} объектов с координатами >1")

        # Показываем первые 3 проблемных объекта
        for detail in problem_details[:3]:
            print(f"     Строка {detail['line']}: {detail['problems']}")

    return {
        'filename': file_path,
        'has_problems': problem_objects > 0,
        'total_objects': total_objects,
        'problem_objects': problem_objects,
        'problem_details': problem_details
    }
```

```python
def remove_files_with_invalid_coordinates(yolo_dir):
    """
    Удаляет txt файлы с координатами >1 и соответствующие изображения
    """

    print("🗑️ УДАЛЕНИЕ ФАЙЛОВ С КООРДИНАТАМИ >1")
    print("═" * 70)

    # Находим все txt файлы
    train_labels = glob.glob(os.path.join(yolo_dir, 'labels', 'train', '*.txt'))
    val_labels = glob.glob(os.path.join(yolo_dir, 'labels', 'val', '*.txt'))
    all_labels = train_labels + val_labels

    invalid_files = []
    deleted_images = []
    deleted_labels = []

    # Проверяем каждый txt файл на наличие координат >1
    for label_file in all_labels:
        has_invalid_coords = check_file_has_invalid_coordinates(label_file)

        if has_invalid_coords:
            invalid_files.append(label_file)

            # Находим и удаляем соответствующее изображение
            base_name = os.path.splitext(os.path.basename(label_file))[0]
            image_deleted = delete_corresponding_image(yolo_dir, base_name)

            if image_deleted:
                deleted_images.append(image_deleted)

            # Удаляем txt файл с невалидными координатами
            os.remove(label_file)
            deleted_labels.append(label_file)
            print(f"🗑️ Удален файл с невалидными координатами: {os.path.basename(label_file)}")

    # Выводим статистику
    print(f"\n📊 ИТОГИ УДАЛЕНИЯ:")
    print("═" * 70)
    print(f"📁 Найдено файлов с координатами >1: {len(invalid_files)}")
    print(f"🖼️ Удалено изображений: {len(deleted_images)}")
    print(f"📝 Удалено txt файлов: {len(deleted_labels)}")

    # Показываем список удаленных файлов
    if deleted_labels:
        print(f"\n📋 УДАЛЕННЫЕ ФАЙЛЫ (первые 10):")
        for label_file in deleted_labels[:10]:
            print(f"   - {os.path.basename(label_file)}")

    return {
        'invalid_files': invalid_files,
        'deleted_images': deleted_images,
        'deleted_labels': deleted_labels
    }

def check_file_has_invalid_coordinates(label_file):
    """
    Проверяет есть ли в файле координаты >1
    """
    try:
        with open(label_file, 'r') as f:
            for line in f:
                line = line.strip()
                if line:
                    parts = line.split()
                    if len(parts) == 5:
                        coords = list(map(float, parts[1:]))
                        if any(coord > 1.0 for coord in coords):
                            return True
        return False
    except Exception as e:
        print(f"❌ Ошибка чтения файла {label_file}: {e}")
        return False

def delete_corresponding_image(yolo_dir, base_name):
    """
    Удаляет изображение соответствующее txt файлу
    """
    image_extensions = ['.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG']

    for split in ['train', 'val']:
        for ext in image_extensions:
            image_path = os.path.join(yolo_dir, 'images', split, base_name + ext)
            if os.path.exists(image_path):
                os.remove(image_path)
                print(f"🗑️ Удалено изображение: {os.path.basename(image_path)}")
                return image_path
    return None
```

```python
def analyze_and_remove_invalid_files(yolo_dir):
    """
    Полный цикл: анализ + удаление файлов с координатами >1
    """

    print("🎯 ПОЛНЫЙ ЦИКЛ: АНАЛИЗ И УДАЛЕНИЕ ФАЙЛОВ С КООРДИНАТАМИ >1")
    print("═" * 70)

    # 1. Сначала анализируем сколько файлов с проблемами
    print("🔍 ЭТАП 1: АНАЛИЗ...")
    analysis_result = analyze_all_yolo_files(yolo_dir)

    if analysis_result['problem_files'] == 0:
        print("🎉 Файлов с координатами >1 не обнаружено!")
        return analysis_result

    print(f"\n⚠️  Обнаружено файлов с координатами >1: {analysis_result['problem_files']}")
    print(f"📊 Проблемных объектов: {analysis_result['problem_objects']}")

    # 2. Запрашиваем подтверждение на удаление
    confirm = input("🗑️  Удалить проблемные файлы и соответствующие изображения? (y/n): ")

    if confirm.lower() != 'y':
        print("❌ Удаление отменено пользователем")
        return analysis_result

    # 3. Выполняем удаление
    print("\n🔧 ЭТАП 2: УДАЛЕНИЕ...")
    removal_result = remove_files_with_invalid_coordinates(yolo_dir)

    # 4. Проверяем результат
    print("\n✅ ЭТАП 3: ПРОВЕРКА...")
    final_check = analyze_all_yolo_files(yolo_dir)

    if final_check['problem_files'] == 0:
        print("🎉 ВСЕ ФАЙЛЫ С КООРДИНАТАМИ >1 УДАЛЕНЫ!")
    else:
        print(f"⚠️  Осталось проблемных файлов: {final_check['problem_files']}")

    return {
        'before_removal': analysis_result,
        'removal_result': removal_result,
        'after_removal': final_check
    }
```

```python
def safe_remove_invalid_files(yolo_dir):
    """
    Безопасное удаление файлов с координатами >1 с созданием бэкапа
    """

    import shutil
    from datetime import datetime

    # Создаем бэкап
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    backup_dir = f"/content/yolo_backup_before_cleaning_{timestamp}"

    print(f"💾 Создаем бэкап: {backup_dir}")
    shutil.copytree(yolo_dir, backup_dir)

    try:
        result = analyze_and_remove_invalid_files(yolo_dir)
        print(f"\n💾 Бэкап сохранен в: {backup_dir}")
        print("💡 Для восстановления выполните: shutil.copytree(backup_dir, yolo_dir)")
        return result

    except Exception as e:
        print(f"❌ Ошибка при удалении: {e}")
        print("🔄 Восстанавливаем из бэкапа...")
        shutil.rmtree(yolo_dir)
        shutil.copytree(backup_dir, yolo_dir)
        print("✅ Восстановление завершено")
        raise e
```

```python
def check_integrity_after_removal(yolo_dir):
    """
    Проверяет целостность датасета после удаления проблемных файлов
    """

    print("🔍 ПРОВЕРКА ЦЕЛОСТНОСТИ ПОСЛЕ УДАЛЕНИЯ")
    print("═" * 70)

    issues = []

    for split in ['train', 'val']:
        images_dir = os.path.join(yolo_dir, 'images', split)
        labels_dir = os.path.join(yolo_dir, 'labels', split)

        # Получаем базовые имена файлов (без расширения)
        image_bases = set()
        for ext in ['.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG']:
            for img_path in glob.glob(os.path.join(images_dir, f"*{ext}")):
                image_bases.add(os.path.splitext(os.path.basename(img_path))[0])

        label_bases = set()
        for label_path in glob.glob(os.path.join(labels_dir, '*.txt')):
            label_bases.add(os.path.splitext(os.path.basename(label_path))[0])

        # Проверяем соответствие
        if image_bases != label_bases:
            images_without_labels = image_bases - label_bases
            labels_without_images = label_bases - image_bases

            if images_without_labels:
                issues.append(f"❌ {split}: {len(images_without_labels)} изображений без разметки")

            if labels_without_images:
                issues.append(f"❌ {split}: {len(labels_without_images)} файлов разметки без изображений")
        else:
            print(f"✅ {split}: все изображения имеют соответствующую разметку")

    if not issues:
        print("🎉 Датсет целостен! Все изображения имеют разметку и наоборот.")
    else:
        print(f"⚠️  Найдено проблем: {len(issues)}")
        for issue in issues:
            print(f"   {issue}")

    return {
        'issues': issues,
        'is_consistent': len(issues) == 0
    }
```

```python
def backup_valid_pairs_only(yolo_dir, backup_name=None):
    """
    Сохраняет на Google Drive только валидные пары (изображение + txt)
    после очистки от координат >1
    """

    from datetime import datetime
    import shutil
    from google.colab import drive

    print("💾 СОХРАНЕНИЕ ВАЛИДНЫХ ПАР (ИЗОБРАЖЕНИЕ + TXT) НА GOOGLE DRIVE")
    print("═" * 70)

    # Монтируем Google Drive
    drive.mount('/content/drive')

    # Создаем имя для бэкапа
    if backup_name is None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_name = f"cow_detection_cleaned_{timestamp}"

    # Путь для сохранения на Google Drive
    drive_backup_path = f'/content/drive/MyDrive/{backup_name}'

    try:
        # Создаем структуру папок на Google Drive
        os.makedirs(os.path.join(drive_backup_path, 'images', 'train'), exist_ok=True)
        os.makedirs(os.path.join(drive_backup_path, 'images', 'val'), exist_ok=True)
        os.makedirs(os.path.join(drive_backup_path, 'labels', 'train'), exist_ok=True)
        os.makedirs(os.path.join(drive_backup_path, 'labels', 'val'), exist_ok=True)

        valid_pairs_count = 0

        print("🔍 Поиск валидных пар файлов...")

        for split in ['train', 'val']:
            images_dir = os.path.join(yolo_dir, 'images', split)
            labels_dir = os.path.join(yolo_dir, 'labels', split)

            if not os.path.exists(images_dir) or not os.path.exists(labels_dir):
                print(f"⚠️  Пропускаем {split}: директории не найдены")
                continue

            # Получаем все базовые имена файлов (без расширения)
            image_bases = set()
            for ext in ['.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG']:
                for img_path in glob.glob(os.path.join(images_dir, f"*{ext}")):
                    base_name = os.path.splitext(os.path.basename(img_path))[0]
                    image_bases.add(base_name)

            label_bases = set()
            for label_path in glob.glob(os.path.join(labels_dir, '*.txt')):
                base_name = os.path.splitext(os.path.basename(label_path))[0]
                label_bases.add(base_name)

            # Находим пересечение - валидные пары
            valid_bases = image_bases.intersection(label_bases)

            print(f"📊 {split}: {len(valid_bases)} валидных пар")

            # Копируем валидные пары
            copied_count = 0
            for base_name in valid_bases:
                # Копируем изображение
                image_copied = False
                for ext in ['.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG']:
                    source_image = os.path.join(images_dir, base_name + ext)
                    if os.path.exists(source_image):
                        target_image = os.path.join(drive_backup_path, 'images', split, base_name + ext)
                        shutil.copy2(source_image, target_image)
                        image_copied = True
                        break

                # Копируем txt файл
                source_label = os.path.join(labels_dir, base_name + '.txt')
                if os.path.exists(source_label):
                    target_label = os.path.join(drive_backup_path, 'labels', split, base_name + '.txt')
                    shutil.copy2(source_label, target_label)

                    if image_copied:
                        copied_count += 1
                        valid_pairs_count += 1

            print(f"   ✅ Скопировано: {copied_count} пар")

        # Копируем конфиг файл если есть
        config_source = os.path.join(yolo_dir, 'cow_dataset.yaml')
        if os.path.exists(config_source):
            shutil.copy2(config_source, os.path.join(drive_backup_path, 'cow_dataset.yaml'))
            print("📄 Конфиг файл скопирован")

        # Создаем обновленный конфиг с новыми путями
        create_updated_config(drive_backup_path, valid_pairs_count)

        # Подсчитываем итоговую статистику
        train_images = len(glob.glob(os.path.join(drive_backup_path, 'images', 'train', '*')))
        train_labels = len(glob.glob(os.path.join(drive_backup_path, 'labels', 'train', '*.txt')))
        val_images = len(glob.glob(os.path.join(drive_backup_path, 'images', 'val', '*')))
        val_labels = len(glob.glob(os.path.join(drive_backup_path, 'labels', 'val', '*.txt')))

        print(f"\n✅ ВАЛИДНЫЕ ПАРЫ СОХРАНЕНЫ НА GOOGLE DRIVE:")
        print(f"   📁 Путь: {drive_backup_path}")
        print(f"   📊 Всего валидных пар: {valid_pairs_count}")
        print(f"   🖼️  Train: {train_images} изображений, {train_labels} разметок")
        print(f"   🖼️  Val: {val_images} изображений, {val_labels} разметок")

        return drive_backup_path

    except Exception as e:
        print(f"❌ Ошибка при сохранении валидных пар: {e}")
        return None

def create_updated_config(backup_path, valid_pairs_count):
    """
    Создает обновленный конфиг файл для сохраненного датасета
    """

    config_content = f"""# YOLO Dataset Configuration - Cleaned Version
# Generated automatically after removing invalid coordinates
# Valid pairs: {valid_pairs_count}

path: {backup_path}
train: images/train
val: images/val

# Number of classes
nc: 1

# Class names
names: ['cow']

# Statistics
# - Removed objects with coordinates >1
# - Only valid image-label pairs preserved
# - Ready for training
"""

    config_path = os.path.join(backup_path, 'dataset_cleaned.yaml')
    with open(config_path, 'w') as f:
        f.write(config_content)

    print(f"📄 Создан обновленный конфиг: {config_path}")
```

```python
import os
import yaml

def create_dataset_config(yolo_dir):
    """
    Создает YAML конфиг файл для датасета
    """

    print("📄 СОЗДАНИЕ КОНФИГ ФАЙЛА ДАТАСЕТА")
    print("═" * 70)

    config_path = os.path.join(yolo_dir, 'cow_dataset.yaml')

    # Проверяем структуру датасета
    print("🔍 ПРОВЕРКА СТРУКТУРЫ ДАТАСЕТА:")

    train_images = os.path.join(yolo_dir, 'images', 'train')
    val_images = os.path.join(yolo_dir, 'images', 'val')
    train_labels = os.path.join(yolo_dir, 'labels', 'train')
    val_labels = os.path.join(yolo_dir, 'labels', 'val')

    # Подсчитываем файлы
    def count_files(directory, extensions):
        count = 0
        for ext in extensions:
            count += len(glob.glob(os.path.join(directory, f"*{ext}")))
        return count

    train_img_count = count_files(train_images, ['.jpg', '.jpeg', '.png'])
    val_img_count = count_files(val_images, ['.jpg', '.jpeg', '.png'])
    train_label_count = len(glob.glob(os.path.join(train_labels, '*.txt')))
    val_label_count = len(glob.glob(os.path.join(val_labels, '*.txt')))

    print(f"   🖼️  Train images: {train_img_count}")
    print(f"   📝 Train labels: {train_label_count}")
    print(f"   🖼️  Val images: {val_img_count}")
    print(f"   📝 Val labels: {val_label_count}")

    # Создаем конфиг
    config_content = {
        'path': yolo_dir,
        'train': 'images/train',
        'val': 'images/val',
        'test': 'images/test',

        'nc': 1,  # Number of classes
        'names': ['cow']  # Class names
    }

    # Сохраняем конфиг
    with open(config_path, 'w') as f:
        yaml.dump(config_content, f, default_flow_style=False, sort_keys=False)

    print(f"✅ Конфиг создан: {config_path}")
    return config_path

def check_dataset_structure(yolo_dir):
    """
    Проверяет структуру датасета
    """

    print("🔍 ПРОВЕРКА СТРУКТУРЫ ДАТАСЕТА")
    print("═" * 70)

    required_dirs = [
        'images/train',
        'images/val',
        'labels/train',
        'labels/val'
    ]

    missing_dirs = []

    for dir_path in required_dirs:
        full_path = os.path.join(yolo_dir, dir_path)
        if not os.path.exists(full_path):
            missing_dirs.append(dir_path)
            print(f"❌ Отсутствует: {dir_path}")
        else:
            file_count = len(os.listdir(full_path))
            print(f"✅ {dir_path}: {file_count} файлов")

    if missing_dirs:
        print(f"\n🚨 Отсутствуют директории: {missing_dirs}")
        return False
    else:
        print("\n🎉 Структура датасета корректна!")
        return True

# Проверяем и создаем конфиг
yolo_directory = '/content/yolo_cow_dataset_saved'

print("🔧 ПОДГОТОВКА ДАТАСЕТА К ОБУЧЕНИЮ")
print("═" * 70)

# 1. Проверяем структуру
structure_ok = check_dataset_structure(yolo_directory)

if structure_ok:
    # 2. Создаем конфиг файл
    config_path = create_dataset_config(yolo_directory)

    # 3. Проверяем что конфиг создался
    if os.path.exists(config_path):
        print(f"\n✅ КОНФИГ СОЗДАН: {config_path}")

        # Показываем содержимое конфига
        with open(config_path, 'r') as f:
            print("📄 СОДЕРЖИМОЕ КОНФИГА:")
            print(f.read())
    else:
        print("❌ Не удалось создать конфиг файл")
else:
    print("❌ Структура датасета некорректна, исправьте перед обучением")
```

```python
def extract_file_id(google_drive_url):
    """
    Извлекает FILE_ID из ссылки Google Drive
    """
    import re

    # Паттерны для разных форматов ссылок
    patterns = [
        r'/file/d/([a-zA-Z0-9_-]+)',
        r'id=([a-zA-Z0-9_-]+)',
        r'folders/([a-zA-Z0-9_-]+)'
    ]

    for pattern in patterns:
        match = re.search(pattern, google_drive_url)
        if match:
            return match.group(1)

    return None

# Пример использования
url = "https://drive.google.com/file/d/1mp3Mt1LCvmYCb-g7Fdvq5Xw1pYSPgtQ2/view?usp=drive_link"
file_id = extract_file_id(url)
print(f"📎 Найден FILE_ID: {file_id}")
```
